```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

ComLitSur = pd.read_csv("ComputerLiteracySurvey.csv")
EntComSur = pd.read_csv("EntryLevelComputingSurvey.csv")
```

### Question 1:
#### How have Students in Entry-Level Computing Courses heard about CCM?
By examining the ways through which prospective students learn about and begin to consider enrolling at CCM, the department can more effectively distribute its recruitment resources.

```python
#Gathering data and creating labels
insight1counts = [EntComSur['heard_ccmwebsite'].value_counts()['Yes'], 
                  EntComSur['heard_socialmedia'].value_counts()['Yes'], 
                  EntComSur['heard_event'].value_counts()['Yes'], 
                  EntComSur['heard_familyfriend'].value_counts()['Yes'], 
                  EntComSur['heard_ccmstudent'].value_counts()['Yes'], 
                  EntComSur['heard_ccmalumni'].value_counts()['Yes'], 
                  EntComSur['heard_hsteacher'].value_counts()['Yes'], 
                  EntComSur['heard_hscounselor'].value_counts()['Yes'], 
                  EntComSur['heard_appadvert'].value_counts()['Yes'], 
                  EntComSur['heard_employer'].value_counts()['Yes'], 
                  EntComSur['heard_billboard'].value_counts()['Yes'], 
                  EntComSur['heard_television'].value_counts()['Yes'], 
                  EntComSur['heard_radio'].value_counts()['Yes'], 
                  EntComSur['heard_other'].value_counts()['Yes']]
insight1labels = ['CCM Website','Social Media','Community Event','Family or Friend','CCM Student','CCM Alumni',
                  'HS Teacher','HS Counselor','In-app Advertisement','Employer','Billboard','Television','Radio','Other']
```

```python
#Creation of the plot
df = pd.DataFrame(dict(labels=insight1labels, counts=insight1counts))
df_sorted = df.sort_values('counts', ascending=False)
plt.title('Reported Sources of Hearing about CCM for Entry-Level Computing Students')
plt.xlabel('Affirmative Responses')
plt.barh('labels', 'counts', data=df_sorted)
```

### Insight 1:
Among methods which can be directly influenced by the department's efforts, students are most likely to learn about CCM through high school staff and through the CCM website. It would be wise for the department to focus on collaboration with local high schools and on maintaining and supplying up-to-date information to the CCM website.

### Question 3:
#### What motivations do students have to take Computer Literacy courses?
Analysis of data regarding students' motivations in taking the courses offered by the department can help determine effective ways to encourage more students to take the courses, like by tailoring the messaging of flyers and other advertisements to fit popular motivations.

```python
#Gathering data and creating labels
insight3counts = [ComLitSur['motivated_required'].value_counts()['Yes'], 
                  ComLitSur['motivated_staycurrenttech'].value_counts()['Yes'], 
                  ComLitSur['motivated_advancecareer'].value_counts()['Yes'], 
                  ComLitSur['motivated_changecareer'].value_counts()['Yes'], 
                  ComLitSur['motivated_developprofession'].value_counts()['Yes'], 
                  ComLitSur['motivated_jobdisplacement'].value_counts()['Yes'], 
                  ComLitSur['motivated_relocation'].value_counts()['Yes'], 
                  ComLitSur['motivated_itcert'].value_counts()['Yes'], 
                  ComLitSur['motivated_financial'].value_counts()['Yes'],
                  ComLitSur['motivated_enrichment'].value_counts()['Yes'],
                  ComLitSur['motivated_curiosity'].value_counts()['Yes'],]
insight3labels = ['Required Class','Maintain Computing Skills','Career Advancement','Career Change',
                  'Professional Development','Job Displacement','Relocation','IT Industry Certifications',
                  'Financial','Personal Enrichment','Curiosity']
```

```python
#Gathering data and creating labels
insight4countshigh = [ComLitSur['experienced_compclass'].value_counts()['High Impact'], 
                      ComLitSur['experienced_compclub'].value_counts()['High Impact'], 
                      ComLitSur['experienced_compcompetition'].value_counts()['High Impact'], 
                      ComLitSur['experienced_compcampschool'].value_counts()['High Impact'], 
                      ComLitSur['experienced_compcampsummer'].value_counts()['High Impact'], 
                      ComLitSur['experienced_compclassap'].value_counts()['High Impact'], 
                      ComLitSur['experienced_compclassdual'].value_counts()['High Impact'], 
                      ComLitSur['experienced_familyfriendinfluence'].value_counts()['High Impact'], 
                      ComLitSur['experienced_familyfriendinfield'].value_counts()['High Impact'], 
                      ComLitSur['experienced_hsstaff'].value_counts()['High Impact'], 
                      ComLitSur['experienced_employerinfluence'].value_counts()['High Impact']]

insight4countssome = [ComLitSur['experienced_compclass'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_compclub'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_compcompetition'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_compcampschool'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_compcampsummer'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_compclassap'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_compclassdual'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_familyfriendinfluence'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_familyfriendinfield'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_hsstaff'].value_counts()['Some Impact'], 
                      ComLitSur['experienced_employerinfluence'].value_counts()['Some Impact']]

insight4countstotal = [ComLitSur['experienced_compclass'].value_counts()['High Impact'] + ComLitSur['experienced_compclass'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_compclub'].value_counts()['High Impact'] + ComLitSur['experienced_compclub'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_compcompetition'].value_counts()['High Impact'] + ComLitSur['experienced_compcompetition'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_compcampschool'].value_counts()['High Impact'] + ComLitSur['experienced_compcampschool'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_compcampsummer'].value_counts()['High Impact'] + ComLitSur['experienced_compcampsummer'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_compclassap'].value_counts()['High Impact'] + ComLitSur['experienced_compclassap'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_compclassdual'].value_counts()['High Impact'] + ComLitSur['experienced_compclassdual'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_familyfriendinfluence'].value_counts()['High Impact'] + ComLitSur['experienced_familyfriendinfluence'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_familyfriendinfield'].value_counts()['High Impact'] + ComLitSur['experienced_familyfriendinfield'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_hsstaff'].value_counts()['High Impact'] + ComLitSur['experienced_hsstaff'].value_counts()['Some Impact'], 
                       ComLitSur['experienced_employerinfluence'].value_counts()['High Impact'] + ComLitSur['experienced_employerinfluence'].value_counts()['Some Impact']]

insight4labels = ['Previous Computing Class','Previous Computing Club','Computing Competition',
                  'Afterschool Computing Camp/Program','Summer Computing Camp/Program','AP Computing Class',
                  'Dual Enrollment Computing Class','Family/Friend Influence','Family/Friend in Computing Field',
                  'HS teacher/counselor Influence','Employer Influence']
```

```python
#Creation of the plot
df = pd.DataFrame(dict(labels=insight4labels, countshigh=insight4countshigh, countssome=insight4countssome, countstotal=insight4countstotal))
df_sorted = df.sort_values('countstotal', ascending=False)
fig, (ax1, ax2) = plt.subplots(2,1, sharex=True)
fig.suptitle('Influence of Experiences to Enroll for Computer Literacy Students')
plt.xlabel('Responses')
ax1.set_title('Some Impact')
ax2.set_title('High Impact')
ax1.barh('labels', 'countssome', data=df_sorted, color='c')
ax2.barh('labels', 'countshigh', data=df_sorted, color='m')
```

### Insight 4:
While the greatest volume of responses cite a previous computing class in high school or middle school as impactful, the highest degree of impact comes from the influence of a family member, a friend, or a counselor. This implies that a sense of shared community and camaraderie could provide a strong pull of students into the Computing department.
